## Display images

```python
import glob
#print(glob.glob("data/train/images/*"))

list_images = (glob.glob("data/train/images/*"))
list_masks = (glob.glob("data/train/masks/*"))
```

```python
group = 1 # specify what group you want to look at - group 1 corresponds to images 0 to 9 (1 to 10 if using 1-indexing)
fig = plt.figure(figsize=(20,20))

for index in np.arange(5*2*(group-1),(group*2-1)*5):
    a = fig.add_subplot(1,5,index % 5 + 1)
    load_image = plt.imread(list_images[index]) 
    imgplot = plt.imshow(load_image)
    load_contours = plt.imread(list_masks[index]) 
    imgcontour = plt.contour(load_contours, colors='r')
    cropped_title = list_images[index][18:]
    a.set_title(cropped_title)

    
for index in np.arange((group*2-1)*5, group*5*2):

    a = fig.add_subplot(2,5,5+(index % 5 + 1))
    load_image = plt.imread(list_images[index]) 
    imgplot = plt.imshow(load_image)
    load_contours = plt.imread(list_masks[index]) 
    imgcontour = plt.contour(load_contours, colors='r')
    cropped_title = list_images[index][18:]
    a.set_title(cropped_title)
```
